```python
 # make bar graph of top 5 features in the data  
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
# file download 
file = "Desktop/worldbank_data_postprocssed.txt" 
df=pd.read_csv(file) 
df.head()
```

```python
# SP.RUR.TOTL.ZG 
df = df.sort_values("SP.RUR.TOTL.ZG")
plt.bar(df['Country'], df['SP.RUR.TOTL.ZG']) 
plt.title=('Country VS SP.RUR.TOTL.ZG ') 
plt.xlabel('Country', fontsize=8)
plt.xticks(rotation=90)
plt.ylabel('SP.RUR.TOTL.ZG')
plt.gcf().set_size_inches(40, 6)
plt.show()
```
